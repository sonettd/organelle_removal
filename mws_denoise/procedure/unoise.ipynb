```python
import glob
```

```python
#trimming and quality filter, already done for MWS dataset, but need to convert to fasta
fps = glob.glob('../input/*.fastq')
for fp in fps:
    fasta_fp = fp.rstrip('.fastq') + '_u.fa'
    fasta_discarded = fp.rstrip('.fastq') + '_u_discarded.fa'
    !/mnt/c/Users/Dylan/Documents/zaneveld/mws_denoise/usearch -fastq_filter $fp -fastq_trunclen 100 -fastq_maxee_rate 0.02 -fastaout $fasta_fp -fastaout_discarded $fasta_discarded
```

```python
fps = glob.glob('../input/*_u.fa')
with open('../input/u_pooled_samples.fa', 'w') as pooled_file:
    for fp in fps:
        with open(fp) as fasta_file:
            for line in fasta_file:
                pooled_file.write(line)
```

```python
!/mnt/c/Users/Dylan/Documents/zaneveld/mws_denoise/usearch -fastx_uniques ../input/u_pooled_samples.fa -fastaout ../input/u_dereplicated.fa -sizeout -relabel Uniq
```
